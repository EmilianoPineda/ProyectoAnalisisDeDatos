```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("hypertension_peru_clean.csv")
df
```

```python
df_filtered = df.drop(columns=['id'])
plt.figure(figsize=(10, 8))
ax = sns.heatmap(df_filtered.corr(), vmin=-1, vmax=1, annot=True, cmap="YlGnBu", linewidths=.5);
```

```python
df_grouped1 = df.groupby('id')[['region', 'hypertension_dx', 'treatment', 'HTA_new']].max()
df_grouped = df.groupby('id')[['hypertension_dx', 'treatment']].max()

df_ball = df_grouped[df_grouped['hypertension_dx'].isin(['Yes', 'No'])]

value_counts_all = df_ball['hypertension_dx'].value_counts()

value_counts_all

fig = plt.figure()
ax = fig.add_subplot()

sns.barplot(value_counts_all.index, value_counts_all, ax=ax, palette='Blues');

ax.set_ylabel('conteo')
ax.set_title('Personas diagnosticadas con hipertension anteriormente', fontsize=13, pad=10);
```

```python
df_grouped['count'] = 1

df_grouped.head()
```

```python
df_ball_with_count = df_grouped[df_grouped['hypertension_dx'].isin(['Yes', 'No'])]

value_counts = df_ball_with_count.groupby(['hypertension_dx', 'treatment'])['count'].sum()

value_counts
```

```python
value_counts = value_counts.unstack(1)

value_counts
```

```python
value_counts = value_counts.fillna(0)
value_counts
```

```python
plt.style.use('seaborn')

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot()

plt1 = ax.bar(value_counts.index, value_counts['Yes'], label='Yes')
plt2 = ax.bar(value_counts.index, value_counts['No'], bottom=value_counts['Yes'])

ax.set_ylabel('count')
ax.set_title('Conteo de frecuencia de personas con hipertension', fontsize=13, pad=15);
plt.legend((plt1[0], plt2[0]), ('Yes', 'No'));
ax.set_ylim(0, 4500);
```
